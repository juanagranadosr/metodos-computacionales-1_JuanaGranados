```python
from IPython.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
from scipy.stats import binom
from scipy.stats import poisson
```

```python
def GetHisto(Data, bins):
    h, bins = np.histogram(Data, bins=bins)
    w = np.diff(bins) # ancho de clase (grosor cada bin)
    m = bins[:-1] + 0.5*w[0] # marca de clase (punto medio en cada bin)
    
    hnorm = h/np.sum(h*w) # frecuencias relativas (porcentajes)
    
    return h, hnorm, m, w
```

## Bernoulli

```python
p = 1/6
N = 1000
```

```python
b = bernoulli.rvs(p, size = N)
```

```python
xbins = np.linspace(0, 2, 3)
```

```python
Hb, HNb, mb, w = GetHisto(b, xbins)
```

```python
plt.bar(mb, HNb, width=w, ec='k')
```

Para le caso de una distribucci√≥n de bernoulli el valor esperado:
$$\mathbb{E}(x) = \sum f(x)x = \frac{1}{6} = p$$

```python
x = np.array([0,1])
E = np.sum(HNb*x)
print(E, 1/6)
```

Para le caso de la varianza:
$$\text{Var}(x) = qp$$

```python
n, p = 2, 1/6
```

```python
binomial = binom.rvs( n,p,size=N )
xbins2 = np.linspace(0,3,4)
xbins2
```
