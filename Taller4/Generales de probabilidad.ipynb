# Ejercicio 4. 

```python
def cumple(n):
    rta = 1
    for i in range(n):
        rta *= ((365-i)/365)
    return rta
```

```python
x = np.linspace(0,80,81)
y = np.zeros(81)
for i in range(80):
    y[i] = cumple(int(x[i]))
plt.scatter(x,y)
plt.xlabel('Número de personas')
plt.ylabel('ℙn')
```

# Ejercicio 9. 

En el ejercicio anterior, imagine que las monedas están truncadas de tal manera que la probabilidad de que la moneda $1$ sea cara es $p_1$ y que sea sello es $1−p_1$. Usando el árbol de probabilidad, ¿cuál es la expresión de la probabilidad de obtener dos caras y dos sellos de este evento? Si el truncamiento de las monedas $1$ y $2$ puede variar como: $0.1 < p_1 < 0.9$ y $0.1 < p:2 < 0.5$, use el árbol de probabilidad para graficar la superficie de probabilidad del evento A. ¿En qué punto la probabilidad es mínima y máxima, y cuáles son sus valores?

```python
p_1 = 0.5 
p_2 = 1-p_1
p_A = 6*(p_1**2)*(p_2**2) #conjunta
print('Probabilidad de obtener dos caras y dos sellos de este evento: ',p_A*100,'%')
```

```python
p_1 = np.linspace(0.1, 0.9, 100)
p_2 = np.linspace(0.1, 0.5, 100)

p1_, p2_ = np.meshgrid(p_1, p_2)
p_A = 6*(p1_**2)*(1-p1_)**2 * (p2_**2)*(1-p2_)**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(p1_, p2_, p_A, cmap= 'coolwarm') #grid
ax.set_xlabel('p_1')
ax.set_ylabel('p_2')
ax.set_zlabel('p(A)')
plt.show()
```
