Un embarque de 10 microchips similares que se env铆a a distribuci贸n tiene 3 aparatos defectuosos. Si una empresa realiza un compra aleatoria de 2 de estos microchips.

```python
# Prueba de que la funci贸n f() es una distribuci贸n de probabilidad
# f(x)>=0
for valor in [0,1,2]:
    display(f"Para x={valor}, f(x):",sym.nsimplify(f_np(valor)))
print("Para todo el dominio los valores de f(x) son no negativos.\n")

# (f(x))=1
c = 0
for k in [0,1,2]:
    c+=f_np(k)
print(f"Como f(x)={c}, la funci贸n f puede ser entonces una funci贸n de densidad de probabilidad.")
```

```python
# Calcular el valor esperado de los microchips defectuosos
n = sym.symbols('n')
series =sym.Sum(x*f, (x, 0, 3) )
display(Math('\hat{\mu}=E(X) =%s' %sym.latex(series)))
print("El valor esperado de microchips defectuosos es:")
display(Math('\hat{\mu}='+sym.latex(series.doit())))
```

# Ejercicio 4
Una caja cu谩ntica tiene 3 electrones, 2 protones y 3 neutrones. Se selecciona una muestra
aleatoria de 4 part铆culas. Si x es el n煤mero de electrones e y es el n煤mero de protones.

```python
x,y = sym.symbols('x y')
f = (sym.binomial(3,x)*sym.binomial(2,y)*sym.binomial(3,4-x-y))/sym.binomial(8,4)
f_np = sym.lambdify((x,y), f, 'numpy')
display(Math('f(x,y)='+ sym.latex(f)))
```

```python
def funcionconjunta(x,y):
    
    num= (sym.binomial(3,x)*sym.binomial(2,y)*sym.binomial(3,4-x-y))
    denom =sym.binomial(8,4)
    return num/denom
```

```python
# Muestre que la distribuci贸n de probabilidad conjunta f(x,y) es la siguiente

# Justificar la distribuci贸n de probabilidad conjunta

# Probar que f(x,y)>=0, para todo el dominio
M=np.zeros([4,3])
for electrones in range(4):
    for protones in range(3):
        if 0<=(4-electrones-protones) and (4-electrones-protones)<=4:
            M[electrones,protones]=funcionconjunta(electrones,protones)
        else:
            M[electrones,protones]=0
            #print(electrones,protones)
display(M)
print(M>=0)
print("Por lo tanto, f(x,y)>0 para todo el dominio.\n")

# Probar (f(x,y))=1
print(f"Ahora calculamos (f(x,y)={np.sum(M)}.")
print("Por lo tanto la funci贸n f(x,y) es una funci贸n de distribuci贸n de probabilidad conjunta v谩lida.")
```
